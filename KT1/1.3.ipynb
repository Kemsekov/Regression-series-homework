```python
import statsmodels.api as sm

ts_A = sm.datasets.get_rdataset("AirPassengers").data["value"].values
ts_B = sm.datasets.get_rdataset("CO2").data["value"].values
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(ts_A)
plt.title("AirPassengers")
plt.subplot(1,2,2)
plt.plot(ts_B)
plt.title("CO2")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

seasonal_decompose(ts_A,model='additive',period=12).plot()
plt.show()
seasonal_decompose(ts_B,model='additive',period=12).plot()
plt.show()
```

```python
from statsmodels.tsa.seasonal import STL
STL(ts_A,period=12).fit().plot()
plt.show()
```
