```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

ts_A = sm.datasets.get_rdataset("AirPassengers").data["value"].values
ts_B = sm.datasets.get_rdataset("CO2").data["value"].values

print(len(ts_A))
print(len(ts_B))

plt.figure(figsize=(12,4))
plt.plot(ts_A)
plt.title("AirPassengers")
plt.grid(True)
plt.show()

plt.figure(figsize=(12,4))
plt.plot(ts_B)
plt.title("CO2 (statsmodels rdataset)")
plt.grid(True)
plt.show()
```

Обе модели точно аддитивные
